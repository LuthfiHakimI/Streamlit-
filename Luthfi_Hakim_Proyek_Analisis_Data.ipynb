## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
#Selanjutnya saya mencoba groupby dari product category name yang memiliki jumlah penjualan terbesar pertama hingga ke 5 untuk menjawab pertanyaan pertama analisis diawal
top_5 = df.groupby(by='product_category_name').size().sort_values(ascending=False).head(5)
top_5.plot(kind='bar', title='Top 5 Kategori Barang Terlaris Selama Penjualan')
plt.xlabel('Kategori Produk')
plt.ylabel('Jumlah Penjualan')
plt.show()
```

### Pertanyaan 2:

```python
# Membuat histogram dari kolom review_score
plt.hist(df['review_score'], bins=5, edgecolor='black')  # Menentukan jumlah bins sesuai kebutuhan
plt.title('Review Score Distribution')
plt.xlabel('Review Score')
plt.ylabel('Frequency')

# Rata-rata review score
average_review_score = df['review_score'].mean()
plt.axvline(x=average_review_score, color='r', linestyle='--', label='Mean Review Score')

# Menampilkan legenda
plt.legend()

# Menampilkan histogram
plt.show()
```

```python
average_review_score = df['review_score'].mean()
print('Average Review Score:', average_review_score)

```

### Pertanyaan 3:

```python
# Filter data untuk kategori barang terlaris (top 5)
top_5_products = df[df['product_category_name'].isin(top_5.index)]

# Menghitung rata-rata review score untuk setiap kategori
top_5_avg_scores = top_5_products.groupby('product_category_name')['review_score'].mean()

# Mencetak rata-rata review score untuk setiap kategori
for category, avg_score in top_5_avg_scores.items():
    print(f"Average Review Score for {category}: {avg_score:.2f}")
```

```python
# Filter data untuk kategori barang selain top 5
other_products = df[~df['product_category_name'].isin(top_5.index)]

# Menghitung rata-rata review score untuk setiap kategori selain top 5
other_avg_scores = other_products.groupby('product_category_name')['review_score'].mean()

# Mencetak rata-rata review score untuk setiap kategori selain top 5
for category, avg_score in other_avg_scores.items():
    print(f"Average Review Score for {category}: {avg_score:.2f}")
```

```python
# Mengatur ukuran grafik
plt.figure(figsize=(10, 6))

# Menggunakan fungsi bar untuk membuat plot batang
plt.bar(top_5_avg_scores.index, top_5_avg_scores.values, color='blue', label='Top 5 Categories')
plt.bar(other_avg_scores.index, other_avg_scores.values, color='orange', label='Other Categories')

# Menambahkan label sumbu-x dan sumbu-y
plt.xlabel('Category')
plt.ylabel('Average Review Score')

# Menambahkan judul grafik
plt.title('Comparison of Average Review Scores for Top 5 Categories vs Other Categories')

# Menampilkan legenda
plt.legend()

# Menampilkan grafik
plt.show()
```

**Detail hanya 5 top terlaris saja**
